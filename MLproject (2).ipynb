```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
#genres	
#id	
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords', 'cast', 'crew']]
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# tranform into ['Action', 'Adventure', 'Fantasy', 'Scify']
```

```python
import ast # it is used to convert data
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval (obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] =movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval (obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break  
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval (obj):
        if i ['job'] == 'Director':
            L.append(i['name'])
            break  
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['crew'] 
```

```python
movies.head()
```

```python
movies['overview'][0] # we have to covert this in list format
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# now we have to remove spaces in the names for example Sam Worthington and Sam Mendes . they have same first name and our system should not get confused 
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " " . join(x))
```

```python
new_df['tags'][0]
```

```python
# make everything in lower case as recommended
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# vectorization
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
         y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
# every movie is vector now . to recommend and find the closest one 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')  
# stop words are the words that are used in sentence but doesn't have main Meaning like a, the, and, to , etc
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```
